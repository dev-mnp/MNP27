## Step 1 - Adding Distribution QTY and Grouping

```python
from importlib.metadata import Distribution
import pandas as pd
```

```python
master_data = pd.read_excel("/Users/aswathshakthi/PycharmProjects/MLMR/MNP26/data/master-entry-all-2026-02-28.xlsx")
master_data = master_data[master_data["Quantity"]!=0].reset_index(drop=True)
master_data = master_data[master_data["Beneficiary Name"] != "Additional"]
master_data["Total Value"] = (master_data["Quantity"] * master_data["Cost Per Unit"]).round(2)
master_data = master_data.reset_index(drop=True)
master_data
```

```python
print(master_data["Total Value"].sum())
print(master_data["Quantity"].sum())
```

```python
master_data.loc[master_data['Beneficiary Type'] != 'District', 'Names'] = master_data['Application Number'] + ' - ' + master_data['Beneficiary Name']
master_data.loc[master_data['Beneficiary Type'] == 'District', 'Names'] = master_data['Beneficiary Name']
# master_data = master_data.drop(columns=["Name"])

cols = list(master_data.columns)
cols.insert(1, cols.pop(cols.index('Names')))  # Move 'Concatenated' to second position
df = master_data[cols]
df
```

```python
df.isna().sum()
df["Aadhar Number"] = df["Aadhar Number"].fillna(0)
df["Handicapped Status"] = df["Handicapped Status"].fillna("NA")
df["Gender"] = df["Gender"].fillna("NA")
df["Gender Category"] = df["Gender Category"].fillna("NA")
df["Comments"] = df["Comments"].fillna("None")
df
```

```python
group_cols = [
    'Application Number',
    'Names',
    'Beneficiary Name',
    'Requested Item',
    'Handicapped Status',
    'Gender',
    'Gender Category',
    'Beneficiary Type',
    'Item Type',
    'Article Category',
    'Super Category Article',
    'Requested Item Tk',
]

grouped = (
    df.groupby(group_cols, as_index=False)
      .agg({"Quantity": "sum", "Total Value": "sum"})
)
grouped
```

```python
## Waiting hall qty + token quantity = Quantity ---> Calcualtion
def check_quantity_balance(
    df: pd.DataFrame,
    qty_col: str = "Quantity",
    waiting_col: str = "Waiting Hall Quantity",
    token_col: str = "Token Quantity",
    fillna_zero: bool = True):
    d = df.copy()

    if fillna_zero:
        d[[qty_col, waiting_col, token_col]] = d[[qty_col, waiting_col, token_col]].fillna(0)

    d["Calculated Quantity"] = d[waiting_col] + d[token_col]
    d["Quantity Match"] = d[qty_col] == d["Calculated Quantity"]
    d["Difference"] = d[qty_col] - d["Calculated Quantity"]

    mismatches = d[~d["Quantity Match"]].copy()

    summary = {
        "total_rows": len(d),
        "matched_rows": int(d["Quantity Match"].sum()),
        "mismatched_rows": int((~d["Quantity Match"]).sum()),
    }

    return summary, mismatches, d
summary, mismatches, checked_df = check_quantity_balance(step2)
print(summary)
mismatches.head()
```
