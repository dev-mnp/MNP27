```python
import pandas as pd
from fillpdf import fillpdfs
import PyPDF2
import os
# Define a function to fill PDF for each row

def get_value_aid(row):
    item = str(row.get("Requested Item", "")).lower()
    is_aid = "aid" in item  # catches "Business Aid", "Auto (3 wheeler) - Aid", etc.

    if not is_aid:
        return ""

    total_value = row.get("Total value", row.get("Total Value", ""))
    if pd.isna(total_value):
        return ""
    return str(total_value)


def fill_pdf_for_row(row):
    data_dict = {
        # 'District': row['District'],
        'address': row['Address'],
        'bf name': row['Names'],
        'App no': row['Application Number'],
        'token': row['Start Token No.'],
        'mobile': row['Mobile'],
        'Aadhar': row['Aadhar'],
        'article': row['Requested Item Tk'],
        'value_aid': get_value_aid(row)   # conditional value
    }
    filled_pdf_path = f"/Users/aswathshakthi/PycharmProjects/MLMR/MNP26/Token/Acknowledgment/result/{row['Start Token No.']}_filled.pdf"  # Output path for each filled PDF
    fillpdfs.write_fillable_pdf('/Users/aswathshakthi/PycharmProjects/MLMR/MNP26/Token/Acknowledgment/data/Public ACK 2026.pdf', filled_pdf_path, data_dict)

    return filled_pdf_path

# Apply the function to each row and get list of filled PDF paths
filled_pdf_paths = data.apply(fill_pdf_for_row, axis=1).tolist()

# Merge all filled PDFs into a single PDF file
merged_pdf_path = "/Users/aswathshakthi/PycharmProjects/MLMR/MNP26/Token/Acknowledgment/result/all_acknowledgment.pdf"
with open(merged_pdf_path, "wb") as output_pdf:
    merger = PyPDF2.PdfMerger()
    for pdf_path in filled_pdf_paths:
        merger.append(pdf_path)
    merger.write(output_pdf)


# # Delete all files in result folder except final merged PDF
# result_dir = "/Users/aswathshakthi/PycharmProjects/MLMR/MNP26/Token/Acknowledgment/result"
# keep_file = os.path.join(result_dir, "all_acknowledgment.pdf")
#
# for file_name in os.listdir(result_dir):
#     file_path = os.path.join(result_dir, file_name)
#     if os.path.isfile(file_path) and os.path.abspath(file_path) != os.path.abspath(keep_file):
#         os.remove(file_path)
```
